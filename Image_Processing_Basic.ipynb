**Image Cropping**

```python
circles.shape
```

```python
patch = (150, 150)
center = (90, 100)

cropped_image = cv2.getRectSubPix(circles, patch, center)
cv2_imshow(cropped_image)
```

```python
def crop(img, center, width, height):
  return cv2.getRectSubPix(img, (width, height), center)

cropped_image = crop(circles, center=(90,100), width=150, height=150)
cv2_imshow(cropped_image)
```

**Rotation**

```python
circles.shape[1]/2, circles.shape[0]/2
```

```python
center
```

```python
center = int(circles.shape[1]/2), int(circles.shape[0]/2)
center

angle = 90
scale = 1.0

M = cv2.getRotationMatrix2D(center, angle, scale)
print(M)

rotated_cat = cv2.warpAffine(circles, M, (center[0]*2, center[1]*2))
cv2_imshow(rotated_cat)
```

***resize***

```python
width, height = int(circles.shape[1]*2), int(circles.shape[0]/2)

resized = cv2.resize(circles, (width, height))
cv2_imshow(resized)
```
