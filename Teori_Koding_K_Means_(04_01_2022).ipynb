### Menentukan Pusat Klaster Awal

```python
# Determine the initial cluster center randomly
import random
def pusat_klaster_awal(k, data):
    X=np.array(data)
    n=len(X)
    list_1=np.arange(0,n)
    pusat_klaster_index = random.choices(list_1, k=k)
    pusat_klaster=np.array(X[pusat_klaster_index])
    return pusat_klaster
```

### Menghitung jarak titik data ke pusat klaster dengan persamaan euclid

```python
def jarak_euclidean(x,y):
    return np.sqrt(pow(x-y, 2))
```

```python
def jarak_centroid2(X,pusat_klaster):
    n=len(X)
    k=np.array(pusat_klaster)
    jarak=[]
    x = list(X)
    for j in range(len(k)):
      jrk = [jarak_euclidean(X, pusat_klaster[j])]
      jarak.append(jrk)
    return np.array(jarak)
```

```python
jarak_centroid = jarak_centroid2(Data,pusat_klaster)
jarak_centroid
```

```python
jarak_centroid = np.transpose(jarak_centroid)
jarak_centroid = jarak_centroid.reshape(len(Data),k)
jarak_centroid
```

### Menentukan Klaster

```python
def menentukan_klaster(jarak_centroid):
    n=len(jarak_centroid)
    minimum_jarak = []
    label_klaster=list()
    for i in range(n):
        max_val, idx = min((val, idx) for (idx, val) in enumerate(jarak_centroid[i]))
        label_klaster.append(idx)
        minimum_jarak.append(max_val)
    return label_klaster
```

```python
def min_jarak(jarak_centroid):
    n = len(jarak_centroid)
    minimum_jarak = []
    label_klaster =list()
    for i in range(n):
        max_val, idx = min((val, idx) for (idx, val) in enumerate(jarak_centroid[i]))
        label_klaster.append(idx)
        minimum_jarak.append(max_val)
    return minimum_jarak
```

### Update Pusat Klaster

```python
def menghitung_pusat_klaster(data, klaster):
    pusat_barudf = dfnew.groupby('Klaster', as_index=False).mean()
    pusat_barudf = pusat_barudf.iloc[::,1:]
    pusat_klaster_baru = np.array(pusat_barudf)
    return pusat_klaster_baru
```

### Cek Perubahan Pusat Klaster

```python
def perubahan_pusat_klaster_euclidean(pusat_klaster_lama, pusat_klaster_baru):
  perubahan = 0
  for a,b in zip(pusat_klaster_lama, pusat_klaster_baru):
    perubahan+=jarak_euclidean(a,b)
  return perubahan
```

```python
FO = perubahan_pusat_klaster_euclidean(pusat_klaster, pusat_klaster_baru)
```
